```python
#Manage data for variable 1
df['Isolate'] = df['H4MH2']
df['Isolate'].value_counts()
```

```python
df['Isolate'] = df['Isolate'].loc[ df['Isolate'] !=' ' ]
df['Isolate'] = df['Isolate'].loc[ df['Isolate'] !='8' ]
df['Isolate'].value_counts()
```

```python
df['Isolate'].replace(['0', '1', '2', '3'] , ['NEVER', 'FEWTIMES', 'SOMETIMES', 'OFTEN'], inplace = True)
df['Isolate'].value_counts()
```

```python
rp.summary_cat(df['Isolate'])
```

```python
#Graphs and tables or statistics
plt.title("How often do you feel isolated from others?")
explode = (0.1, 0.1, 0.2, 0.2, 0.3)
df['Days_missed'].value_counts().plot(kind ='pie', autopct='%1.1f%%', explode=explode, shadow=True, startangle=10, colors=['lavenderblush', 'lightpink', 'pink','palevioletred'])
```

```python
plt.xlabel("Isolate")
plt.title("How often do you feel isolated from others?")
df['Isolate'].value_counts(normalize = True).plot(kind ='barh', color = 'palevioletred', width = 0.8, alpha = 0.5)
```

```python
#Manage data for variable 1
df['Mood_swings'] = df['H4PE4']
df['Mood_swings'].value_counts()
```

```python
df['Mood_swings'] = df['Mood_swings'].loc[ df['Mood_swings'] !=' ' ]
df['Mood_swings'] = df['Mood_swings'].loc[ df['Mood_swings'] !='6' ]
df['Mood_swings'] = df['Mood_swings'].loc[ df['Mood_swings'] !='8' ]
df['Mood_swings'].value_counts()
```

```python
df['Mood_swings'].replace(['1', '2', '3', '4', '5'] , ['STRONGLYAGREE', 'AGREE', 'NEUTRAL', 'DISAGREE', 'STRONGLYDISAGREE'], inplace = True)
df['Mood_swings'].value_counts()
```

```python
#Graphs and tables or statistics
plt.title("I have frequent mood swings")
explode = (0.1, 0.1, 0.2, 0.2, 0.3)
df['Mood_swings'].value_counts().plot(kind ='pie', autopct='%1.1f%%', explode=explode, shadow=True, startangle=10, colors=['thistle', 'plum', 'mediumorchid', 'mediumpurple', 'violet'])
```

```python
plt.xlabel("Mood_swings")
plt.title("I have frequent mood swings")
df['Mood_swings'].value_counts(normalize = True).plot(kind ='barh', color = 'thistle', width = 0.8, alpha = 0.5)
```
